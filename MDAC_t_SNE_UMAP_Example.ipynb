### **Example of t-SNE and UMAP for dimensionality reduction using Python**

Based on: https://github.com/lmcinnes/umap

**Import required packages**

```
import umap
from sklearn.manifold import TSNE
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(context="paper", style="white")

import time
```

**Import data set**

```
mnist = fetch_openml("mnist_784", version=1, as_frame=True)
```

```
mnist_sample = mnist.frame.sample(n=5000)
```

**Example image**

MnistExamples.png


**Perform dimensionality reduction using t-SNE and create chart**

```
start = time.time()

emb_tsne = TSNE(random_state=42).fit_transform(mnist_sample.iloc[:,0:784])

end = time.time()
print(end - start)

fig, ax = plt.subplots(figsize=(12, 10))
color = mnist_sample.iloc[:,-1].astype(int)
plt.scatter(emb_tsne[:, 0], emb_tsne[:, 1], c=color, cmap="Spectral", s=10)
plt.setp(ax, xticks=[], yticks=[])
plt.title("MNIST data embedded into two dimensions by t-SNE", fontsize=18)

plt.show()
```

**Perform dimensionality reduction using UMAP and create chart**

```
start = time.time()

reducer = umap.UMAP(random_state=42)
emb_umap = reducer.fit_transform(mnist_sample.iloc[:,0:784])

end = time.time()
print(end - start)

fig, ax = plt.subplots(figsize=(12, 10))
color = mnist_sample.iloc[:,-1].astype(int)
plt.scatter(emb_umap[:, 0], emb_umap[:, 1], c=color, cmap="Spectral", s=10)
plt.setp(ax, xticks=[], yticks=[])
plt.title("MNIST data embedded into two dimensions by UMAP", fontsize=18)

plt.show()
```
